```python
**Finding the outliers**

- We already know that outliers less than Q1-1.5*IQR and Q3+1.5*IQR

- Step-1: 
    Calculate Q1 = 25p
    Calculate Q2= 50p
    Calculate Q3  =75p
    
- Step-2:
    -Calculate IQR = Q3-Q1
    
- Step-3:
    - Calculate LB = Q1-1.5*IQR
    - Calculate UB = Q3+1.5*IQR
- Step-4:
    
    - Con1 = wage_data < LB
    
    - Con2 = wage_data > UB
    
    - Con3  = con1 or-and con2 (use bitwise operator)
    
    step-5: Data[con3]
```

```python
wage_data = visa_df['prevailing_wage']
# Step 1: Calculate Q1, Q2, Q3
Q1 = np.percentile(wage_data,25)
Q2 = np.percentile(wage_data,50)  
Q3 = np.percentile(wage_data,75)

# Step 2: Calculate IQR
IQR = Q3 - Q1

# Step 3: Calculate LB (Lower Bound) and UB (Upper Bound)
LB = Q1 - 1.5 * IQR
UB = Q3 + 1.5 * IQR

# Step 4: Define conditions for outliers
Con1 = wage_data < LB
Con2 = wage_data > UB
Con3 = Con1 | Con2 
# Bitwise operator used to compare 
# Bit data === 0  and 1 ===True false

# Step 5: Extract outliers
outliers_data = wage_data[Con3]
outliers_data
```

**How to deal outliers**

- 1. We can drop the outliers

    - Dropping outliers is an easy task but not recommended
    
    - If you drop the outliers we are dropping the rows 
    
    - Which means we are dropping the other column values also
    
    - May be some imp info we might lose from other column values
    
    - But in industry if outliers percentage less than or equal to 2% of total data, we can drop
    
    
   - For Ex: the no. of outliers in pwage is 427
   
   - The no. of total observations are 25480
   
  - The percentage is 1.67 which is leaa than 2%
  
  
 2. Fill with Median Value
 
 - We know that outliers doesnot affect by Medians
 
 - So its best practise to fill outliers with Median
 
 3. Capping or Flooring
 
 - Instead of fill with Median we can fill with certain values
 
 - For Ex: the outliers which are less than LB fill with LB values
 
 - The outliers more than UB fill with UB values
