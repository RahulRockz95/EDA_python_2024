```python
# Step-1: Read/load the packages
from sklearn.preprocessing import LabelEncoder

#Step-2 save the package /call the method
le = LabelEncoder()

# Step-3: Apply fit transform on data
visa_df['case_status'] = le.fit_transform(visa_df['case_status'])

visa_df
```

**Fit and Transform**

- Fit means develop the Logic

- Transform means process the Logic

- If we observe in map method first we developed dictionary logic

- Then we apply that dictionary to column 

- Developing the dictionary is a kind of a fitting

- Apply the dictionary to column menas we are transforming the data from cat to num

- Whenever we are transforming or changing the data, then use **fit transform**

- If we are not changing the data then only **fit**

**np.where**

- using np.where we can change the categorical data to num

- But np.where only for binary condition

- np.where is same as if-else

- if condition is True: will apply True value
    
- if condition is False: will apply False value
    
- So np.where use for only binary values

- A column should have only 2 unique values

```python
# value == 'certified' replace with 0
# otherwise 1
con = visa_df['education_of_employee'] == 'Certified'
visa_df['case_status']=np.where(con,0,1)
visa_df

# np.where only for binary
```

**One hot encoder**

- One Hot means one will be ON, another will be OFF

- ON represents 1

- OFF represents 0

- For Ex: case_status has 2 unique labels (certified , denied)
    
    - Certified
    
    - Denied
    
- One hot encoder creates new columns which is equal to nu. of unique labels

- For ex: case_status will create 2 new extra columns
    
    - case_status_Certified
    
    - Case_status_Denied
